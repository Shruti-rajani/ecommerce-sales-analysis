```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:/pd/superstore.csv', encoding='latin1')
print("Dataset loaded:" , df.shape)
```

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
profit_margin = (total_profit / total_sales)*100
total_orders = df['Order ID'].nunique()

print('=' *40 )
print('SUPERSTORE BUSSINESS SUMMARY')
print('=' * 40)
print(f"Total Sales:    ${total_sales: ,.2f}")
print(f"Total Profit:   ${total_profit: ,.2f}")
print(f"profit_margin:  {profit_margin:.1f} ")
print(f"Total Orders:   {total_orders}")
print("=" * 40)
```

```python
losing = df.groupby('Sub-Category')["Profit"].sum()
losing = losing[losing < 0].sort_values()

print("Sub-Categories LOSING money: ")
print(losing)
```

```python
state_profit = df.groupby('State')["Profit"].sum().sort_values()

print( "=" * 40)
print("3 WORST states (losing monney): ")
print( "=" * 40)
print(state_profit.head(3))
print( "=" * 40)
print("3 BEST states (most profitable): ")
print( "=" * 40)
print(state_profit.tail(3))
```

## Key Finding & Business Recommenddations

### What I found:
1. **Techonology** is the most profitable category($145,454 profit)
2. **Furniture** has high sales but very low profit, mnearly breaking even
3. **Tables, Bookcases and supplines** are losiung money despite being sold regulary
4. **High discounts (above 20%) consistently lead to losses**
5. **west region** has the highest sales, **south region** the lowest
6. **Consumer segment** drives the most revenue (50% + of total sales)


### Business Recommendations:
1. Stop giving discounts above 20% — the data clearly shows it causes losses
2. Review pricing strategy for Tables and Bookcases — currently selling at a loss
3. Invest more in Technology products — highest profit margin
4. Focus marketing efforts on West and East regions — highest returns
