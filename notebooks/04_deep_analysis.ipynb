```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C:/pd/superstore.csv", encoding='latin1')

df.head()
```

```python
subcat_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values()

plt.figure(figsize=(8,6))
colors =['red' if x < 0 else 'steelblue' for x in subcat_profit.values]
plt.barh(subcat_profit.index, subcat_profit.values, color=colors)
plt.title('Total Sub-Category Profit')
plt.xlabel('Sub-Category')
plt.axvline(x=0, color='black', linewidth = 0.8)
plt.tight_layout()
plt.savefig('subcat_profit.png')
plt.show()
```

```python
subcat_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(6,4))
sns.barplot(x=subcat_profit.index, y=subcat_profit.values,  hue=subcat_profit.index, palette='Greens')
plt.title('Top 5 Most Profitable Sub-Category')
plt.xlabel('Sub-category')
plt.ylabel('Profit')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Discount', y='Profit', alpha=0.4, color='steelblue')
plt.title("Discount vs Profit")
plt.xlabel('Discount Given')
plt.ylabel('Profit Made')
plt.axhline(y=0, color='red', linewidth=0.8)
plt.tight_layout()
plt.savefig('discount_vs_profit.png')
plt.show()
```

```python
state_profit = df.groupby('State')['Profit'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=state_profit.values, y=state_profit.index, hue = state_profit.index, palette = 'Greens', legend=False)
plt.title('Top 10 Most Profitable State')
plt.xlabel('Profit')
plt.ylabel('State')
plt.tight_layout()
plt.savefig('state_profit.png')
plt.show()          
```

```python
state_profit_all = df.groupby('State')['Profit'].sum().sort_values().head(5)

plt.figure(figsize=(7,6))
sns.barplot(x=state_profit_all.values, y=state_profit_all.index, hue=state_profit_all.index, palette='Reds', legend=False)
plt.title('Bottom 5 Least Profitable State')
plt.xlabel("Total Profit ($)")
plt.ylabel('State')
plt.tight_layout()
plt.show()
```
