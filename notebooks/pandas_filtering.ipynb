```python
import pandas as pd

# Load the dataset into a DataFrame
df = pd.read_csv("C:/pd/superstore.csv", encoding='latin1')

# Display first 5 rows to understand the data
df.head()
```

```python
# Filter orders where Region is 'West region'
west_region_orders = df[df['Region'] == 'West']
print(west_region_orders.shape)
```

```python
# Filter rows where Discount is greater than 0.5 (high discount orders)
losse = df[df['Profit'] < 0]
print(losse.shape)
```

```python
discount = df[df['Discount'] > 0.5]

# Filter rows where:
# Category is 'Furniture'
# AND Profit is less than 0 (loss-making furniture orders)
print(discount.shape)
```

```python
profit_furniture = df[(df['Category'] == 'Furniture') & (df['Profit'] < 0)]

# Print shape → (714, 21)
# This means:
# 714 rows = 714 loss-making furniture orders
# 21 columns = total columns in dataset
print(profit_furniture.shape)
```

```python
bookcases_loss = df[df['Sub-Category'] == 'Bookcases']['Profit'].sum()

print("Total Profit/Loss for Bookcases:", bookcases_loss)
```
